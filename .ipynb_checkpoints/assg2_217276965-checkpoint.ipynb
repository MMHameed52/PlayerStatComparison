```python
# Name: [Mohammad Mustafaa Hameed]

# Email: [autipre@gmail.com]

# Student ID: [217276965]

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

df = pd.read_csv('players_stats_by_season_full_details.csv')

df['PPG'] = df['PTS'] / df['GP']      
df['MPG'] = df['MIN'] / df['GP']      
df['TOVPG'] = df['TOV'] / df['GP']    
df['3PAPG'] = df['3PA'] / df['GP']    
df['3PMPG'] = df['3PM'] / df['GP']    
df['FGAPG'] = df['FGA'] / df['GP']    
df['FGMPG'] = df['FGM'] / df['GP']    
df['FPG'] = df['PF'] / df['GP']       

unique_leagues = df['League'].unique()
print(unique_leagues)

nba_df = df[df['League'] == 'NBA']
print(nba_df.head())

grouped_df = df.groupby(['Player', 'Season']).agg({
    'PTS': 'sum',
    'FGA': 'sum',
    'FTA': 'sum'
}).reset_index()

grouped_df['TotalAttempts'] = grouped_df['FGA'] + grouped_df['FTA']
threshold_df = grouped_df[grouped_df['TotalAttempts'] >= 1000].copy()

def calculate_ts(row):
    ts_percent = row['PTS'] / (2 * (row['FGA'] + 0.44 * row['FTA']))
    return ts_percent

threshold_df.loc[:, 'TS%'] = threshold_df.apply(calculate_ts, axis=1)
sorted_df = threshold_df.sort_values(by='TS%', ascending=False)
print(sorted_df[['Player', 'Season', 'TS%']].head(10))

mean_birth_year = df['birth_year'].mean()
mean_weight_kg = df['weight_kg'].mean()
mean_height_cm = df['height_cm'].mean()

df['birth_year'] = df['birth_year'].fillna(mean_birth_year)
df['weight_kg'] = df['weight_kg'].fillna(mean_weight_kg)
df['height_cm'] = df['height_cm'].fillna(mean_height_cm)

print(df[['birth_year', 'weight_kg', 'height_cm']].head())

def calculate_age(row):
    try:
        season_start_year = int(row['Season'][0:4])  
        birth_year = int(row['birth_year'])
        return season_start_year - birth_year
    except ValueError:
        return None  

df['Age'] = df.apply(calculate_age, axis=1)
print(df[['Player', 'Season', 'birth_year', 'Age']].head())

df.drop('birth_year', axis=1, inplace=True)
print(df.head())

nba_regular_df = df.loc[(df['League'] == 'NBA') & (df['Stage'] == 'Regular_Season')]
df = nba_regular_df.copy()
print(df.head())

grouped_data = df.groupby(['Player', 'Season']).agg({
    'PTS': 'sum',
    'FGA': 'sum',
    'FTA': 'sum'
}).reset_index()

filtered_data = grouped_data[(grouped_data['FGA'] + grouped_data['FTA']) >= 1000].copy()

def calculate_ts(row):
    points = row['PTS']
    fga = row['FGA']
    fta = row['FTA']

    denominator = 2 * (fga + 0.44 * fta)
    if denominator == 0:
        return None  
    else:
        ts_percentage = points / denominator
        return ts_percentage

filtered_data.loc[:, 'TS%'] = filtered_data.apply(calculate_ts, axis=1)

player_season_stats = filtered_data[['Player', 'Season', 'PTS', 'FGA', 'FTA', 'TS%']].copy()
print(player_season_stats.head())

sorted_stats = player_season_stats.sort_values(by='TS%', ascending=False)
top_shooters = sorted_stats.head(10)  
print(top_shooters)
#The Top Scorers are:
#Rudy Gobert (2018 - 2019): TS% of 0.681760
#Rudy Gobert (2016 - 2017): TS% of 0.681295
#Stephen Curry (2017 - 2018): TS% of 0.675106
#DeAndre Jordan (2016 - 2017): TS% of 0.673429
#Stephen Curry (2015 - 2016): TS% of 0.669391
#Maybyner Nene (2010 - 2011): TS% of 0.656533
#Amar'e Stoudemire (2007 - 2008): TS% of 0.656418
#Steve Nash (2006 - 2007): TS% of 0.653897
#Brent Barry (2001 - 2002): TS% of 0.651765
#Kevin Durant (2016 - 2017): TS% of 0.650649

#Stephen Curryâ€™s TS% in these seasons reflects not just his ability to score but to do so with remarkable efficiency, emphasizing his status as one of the best shooters in the history of basketball. 

bc_seasons = [f"{year} - {year+1}" for year in range(1999, 2009)]
ac_seasons = [f"{year} - {year+1}" for year in range(2009, 2023)]
BC = df.loc[df['Season'].isin(bc_seasons)]
AC = df.loc[df['Season'].isin(ac_seasons)]

print("BC DataFrame:")
print(BC.head())

print("\nAC DataFrame:")
print(AC.head())


fig, axes = plt.subplots(3, 1, figsize=(10, 15))
titles = ['Height Distribution', 'Weight Distribution', 'Age Distribution']
columns = ['height_cm', 'weight_kg', 'Age']

for ax, title, column in zip(axes, titles, columns):
    ax.hist(BC[column].dropna(), bins=30, alpha=0.5, label='Before Curry (BC)', color='blue')
    ax.hist(AC[column].dropna(), bins=30, alpha=0.5, label='After Curry (AC)', color='red')

    ax.set_title(title)
    ax.set_xlabel(column)
    ax.set_ylabel('Count')
    ax.legend()

plt.tight_layout()
plt.show()

# The height and age distribution is very similar 
# The players weigh a little more post Curry

BC.loc[:, 'MPG'] = BC['MIN'] / BC['GP']
AC.loc[:, 'MPG'] = AC['MIN'] / AC['GP']
fig, axes = plt.subplots(2, 1, figsize=(10, 10))  

configs = [
    ('GP', 'Number of Games Played'),
    ('MPG', 'Average Minutes Per Game')
]

for ax, (column, title) in zip(axes, configs):
    ax.hist(BC[column].dropna(), bins=30, alpha=0.5, label='Before Curry (BC)', color='blue')
    ax.hist(AC[column].dropna(), bins=30, alpha=0.5, label='After Curry (AC)', color='red')
    ax.set_title(title)
    ax.set_xlabel(column)
    ax.set_ylabel('Count')
    ax.legend()

plt.tight_layout()
plt.show()

# While the number of games played is generally the same the BC players do have a small amount that played more games.
# There are a much higher number of players that played longer than average minutes BC
#It does seem based on the graphs that BC players played more games and minutes.

fig, axes = plt.subplots(3, 1, figsize=(10, 15))

metrics = ['3PAPG', 'TOVPG', 'FPG']
titles = ['3-Point Attempts Per Game', 'Turnovers Per Game', 'Fouls Per Game']

for ax, metric, title in zip(axes, metrics, titles):
    ax.hist(BC[metric].dropna(), bins=30, alpha=0.5, label='Before Curry (BC)', color='blue')
    ax.hist(AC[metric].dropna(), bins=30, alpha=0.5, label='After Curry (AC)', color='red')
    ax.set_title(title)
    ax.set_xlabel(metric)
    ax.set_ylabel('Count')
    ax.legend()

plt.tight_layout()
plt.show()



_, p_3papg = ttest_ind(BC['3PAPG'].dropna(), AC['3PAPG'].dropna(), equal_var=False)
print(f"3PAPG T-test p-value: {p_3papg:.4f}")

_, p_tovpg = ttest_ind(BC['TOVPG'].dropna(), AC['TOVPG'].dropna(), equal_var=False)
print(f"TOVPG T-test p-value: {p_tovpg:.4f}")

_, p_fpg = ttest_ind(BC['FPG'].dropna(), AC['FPG'].dropna(), equal_var=False)
print(f"FPG T-test p-value: {p_fpg:.4f}")
#Receiving a p-value of 0 in t-tests suggests that the differences in the means of 3PAPG, TOVPG, FPG between BC and AC periods are extremely statistically significant. Indicating very strong evidence against the null hypothesis (which states that there is no difference between the means of the two groups).

# 3PAPG: An extremely low p-value here indicates a significant change in the frequency of 3-point attempts across the league. Given the context (Stephen Curry's influence), this likely reflects an increased emphasis on 3-point shooting in the AC era.
#TOVPG: A significant change in the turnover rates could reflect changes in game pace, playing style, or offensive strategies.
#FPG: Changes in fouls per game could suggest alterations in defensive strategies or changes in officiating and rules that influence how games are refereed.

# It's hard to say whether Steph Curry has changed the game based on the data provided. There could be other factors that spreadsheet cannot show. I'm not surprised that things have changed because it's a very large time period and games are always evolving.

Steph = df[df['Player'] == 'Stephen Curry']
print(Steph.head())

fig, ax = plt.subplots(1, 2, figsize=(14, 7))


ax[0].scatter(AC['3PMPG'], AC['3PAPG'], color='blue', alpha=0.5, label='All Players (AC)')
ax[0].scatter(Steph['3PMPG'], Steph['3PAPG'], color='red', label='Stephen Curry', marker='^', s=100) 
ax[0].set_xlabel('3-Point Makes Per Game (3PMPG)')
ax[0].set_ylabel('3-Point Attempts Per Game (3PAPG)')
ax[0].set_title('3-Point Shot Metrics')
ax[0].legend()

ax[1].scatter(AC['FGMPG'], AC['FGAPG'], color='blue', alpha=0.5, label='All Players (AC)')
ax[1].scatter(Steph['FGMPG'], Steph['FGAPG'], color='red', label='Stephen Curry', marker='^', s=100)  
ax[1].set_xlabel('Field Goals Made Per Game (FGMPG)')
ax[1].set_ylabel('Field Goals Attempts Per Game (FGAPG)')
ax[1].set_title('Field Goal Metrics')
ax[1].legend()

plt.tight_layout()
plt.show()
```
